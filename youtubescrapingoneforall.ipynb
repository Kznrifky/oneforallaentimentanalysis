```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time
```

```python
TARGET_KOMENTAR = 2000
URL_VIDEO = 'https://www.youtube.com/watch?v=6jdG2WQEgyI'

options = Options()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--blink-settings=imagesEnabled=false')

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

print("Browser (WebDriver) berhasil dijalankan.")
```

```python
print(f"Mengakses URL: {URL_VIDEO}")
driver.get(URL_VIDEO)

print("Menunggu halaman video dimuat...")
time.sleep(5)

komentar_unik = set()

while len(komentar_unik) < TARGET_KOMENTAR:
    last_height = driver.execute_script("return document.documentElement.scrollHeight")

    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")

    time.sleep(2.5)

    authors = driver.find_elements(By.CSS_SELECTOR, '[id="author-text"]')
    comments = driver.find_elements(By.CSS_SELECTOR, '[id="content-text"]')

    for i in range(len(authors)):
        try:
            data_tuple = (authors[i].text, comments[i].text)
            if data_tuple not in komentar_unik:
                 komentar_unik.add(data_tuple)
        except IndexError:
            continue

    print(f"Progress: {len(komentar_unik)} / {TARGET_KOMENTAR} komentar unik ditemukan...")

    new_height = driver.execute_script("return document.documentElement.scrollHeight")
    if new_height == last_height:
        print("Telah mencapai akhir halaman. Berhenti scrolling.")
        break

print(f"\nScraping selesai. Total {len(komentar_unik)} komentar unik berhasil dikumpulkan.")
```

```python
list_komentar_final = []
for data in komentar_unik:
    list_komentar_final.append({
        'Penulis': data[0],
        'Komentar': data[1]
    })

df = pd.DataFrame(list_komentar_final)
nama_file = 'animasi.csv'
df.to_csv(nama_file, index=False, encoding='utf-8-sig')

print(f"Data komentar berhasil disimpan ke file '{nama_file}'")
```
